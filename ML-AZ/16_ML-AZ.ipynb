```python
# Apriori Algorithm

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
rawdata = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
print(rawdata)
```

```python
transactions = []
for i in range(0, 7501):
    transactions.append([str(rawdata.values[i, j]) for j in range(0, 20)])
```

```python
# Training Apriori on the dataset
from apyori import apriori
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_lenght = 2)
```

```python
# Visualising the results
results = list(rules)
```
